```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = mpimg.imread('dog.jpeg')
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image, cmap='Greys_r')
```

```python
inverted_blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
f, (ax1) = plt.subplots(1, figsize=(20,10))
ax1.set_title('inverted blurred image')
ax1.imshow(inverted_blurred, cmap='Greys_r')
```

```python
final = cv2.divide(gray_image, 255 - inverted_blurred, scale=240)
plt.figure(figsize=(15,15))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
plt.figure(figsize=(30,30))
plt.subplot(1,5,1)
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
